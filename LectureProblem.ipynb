<b> Part 2: Hamiltonian and Gate creation
    
The cost function which we try to minimize is given by:
    
$C(x) = \sum_{i,j} w_{i,j}x_i \left( 1-x_j \right)$ for $ x_{i,j} \in \{0,1\}$
    
,where $w_{i,j}$ accounts for the possibility of different weights for the edges (See Homework).
    
Now we need to translate the cost function into a Hamiltonian for which the groundstate encodes the optimal solution. For that we can use the Pauli Z operator on the computainonal basis $x_i \to \frac{1-Z_i}{2}$.
How does the corresponding Hamiltonian look like?

$H_C = \sum_{i<j}\frac{1}{2}w_{ij}(1-Z_i Z_j) + \mathbf{const}$

Lets find the solution for the simple graph above using QAOA.

```python
G = nx.Graph()
G.add_nodes_from([0, 1, 2, 3])
G.add_edges_from([(0, 1), (1, 2),(2, 3), (3, 0)])
pos=nx.spring_layout(G,seed = 1)
nx.draw(G, with_labels=True, alpha=0.8, node_size=500, pos = pos)
```

```python
qaoa_circuit = QuantumCircuit(len(G.nodes()))
```

In order to act with our $H_c$ on our circuit is to express it as a unitary gate. Lets see how the ZZ Operator acts on a two qubit state. $e^{i\gamma Z\otimes Z}|ab\rangle = e^{-i (-1)^{a\bigoplus b} \gamma} |ab\rangle$. Consider the follwing gate:

```python
qc = QuantumCircuit (2)
gamma = Parameter ('gamma')
qc.rzz(2*gamma, 0, 1)
qc.decompose().draw()
```

We can verify that this sequence of gates does indeed produce the desired state:

$|a\rangle|a\bigoplus b\rangle$ 
<br> $|a\rangle|e^{-i (-1)^{a\bigoplus b} \gamma} a\bigoplus b\rangle$
<br> $|a\rangle|e^{-i (-1)^{a\bigoplus b} \gamma} a\bigoplus b\rangle \bigoplus a  = e^{-i (-1)^{a\bigoplus b} \gamma} |ab\rangle $.

<b> Part 3: Circuit
<br>
We will now build the quantum circuit step by step:
For a Graph with $n$ nodes we need a quantum register with $n$ qubits. We start by preparing the initial state in a totaly mixed superpositon. This state is an eigenstate of the Mixer hamiltonian $H_B$.

```python
def initial_state(circuit):
    for i in range(circuit.num_qubits):
        circuit.h(i)
    return circuit

qaoa_circuit = initial_state(qaoa_circuit)
qaoa_circuit.draw()
```

```python
def problem_unitary(graph,circuit,gamma):
    for edge in list(G.edges()):
        circuit.rzz(2*gamma,edge[0],edge[1])
    return circuit
qaoa_circuit.barrier()
qaoa_circuit = problem_unitary(G, qaoa_circuit, gamma)
qaoa_circuit.decompose().draw()
```

The next step of the QAOA algorithm is the unitary matrix given by the Mixer Hamiltonian. It is simply a rotation around the X-axis by the angle $2\beta$ for all qubits.

```python
beta = Parameter('beta')
def mix_unitary(circuit,beta):
    for i in range(circuit.num_qubits):
        circuit.rx(2*beta,i)
    return circuit
qaoa_circuit.barrier()
qaoa_circuit = mix_unitary(qaoa_circuit, beta)
qaoa_circuit.draw()
```

We have now created one cycle of the trotterization in the QAOA algorithm (p=1). Next we generalize this to a circuit of p iterations of the Problem and Mixing Unitary. At the end of the circuit we measure all qubits in the computational basis.

```python
def create_qaoa_circuit(graph, theta):
    p = len(theta)//2
    gamma = theta[p:]
    beta = theta[:p]
    qaoa_p = QuantumCircuit(len(graph.nodes()))
    qaoa_p = initial_state(qaoa_p)
    for i in range(p):
        qaoa_p = problem_unitary(graph, qaoa_p, gamma[i])
        qaoa_p = mix_unitary(qaoa_p, beta[i])
    qaoa_p.measure_all()
    return qaoa_p
    
    
```

Using the results from the quantum circuit we calculate the cost function. Let's code the cost function of our graph.

```python
def get_cost(graph, state):
    cost = 0
    for i,j in graph.edges():
        if state[i] != state[j]:
            cost -= 1
    
    return cost
```

We take multiple shots of the quantum circuit and calculate the expectation value of the cost function with this set of parameters. Then we can use a classical optimizer to minimize the expecation value and find the optimal state that solves the Max Cut problem.

```python
def expectation_value(graph, counts):
    avg = 0
    sum_count = 0
    for state, count in counts.items():
        cost = get_cost(graph, state)
        avg += cost*count
        sum_count += count
        
    return avg/sum_count
```

Let's put it all togteher to create a function which we can classically optimize.

```python
def run_qpu(graph, shots=512):
    backend = Aer.get_backend('aer_simulator')
    backend.shots = shots
    
    def run_circuit(theta):
        qc = create_qaoa_circuit(graph, theta)
        counts = backend.run(qc, seed_simulator = 5, nshots= shots).result().get_counts()
        return expectation_value(graph,counts)
    
    return run_circuit
```

```python
from scipy.optimize import minimize

p=1
theta = [1.0 for i in range(2*p)]
exp_value = run_qpu(G)

res = minimize(exp_value, theta , method = 'COBYLA')
res
```

<b> Analyzing the Results:
    
    We can now take our optimized parameters and run the circuit again:

```python
from qiskit.visualization import plot_histogram

backend = Aer.get_backend('aer_simulator')
backend.shots = 512

qc_res = create_qaoa_circuit(G, res.x)

counts = backend.run(qc_res, seed_simulator = 5).result().get_counts()

plot_histogram(counts)
```
